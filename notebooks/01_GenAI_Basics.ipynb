```python
# Install tokenization library
!pip install tiktoken --quiet

import tiktoken

# 1. Initialize the tokenizer (GPT-4 uses 'cl100k_base')
encoding = tiktoken.get_encoding("cl100k_base")

text = "Generative AI is transforming the industry in 2026!"

# 2. Convert text to tokens
tokens = encoding.encode(text)

print(f"Original Text: {text}")
print(f"Tokens (IDs): {tokens}")
print(f"Total Tokens: {len(tokens)}")

# 3. Decode tokens back to strings to see the 'splits'
print("\n--- Token Breakdown ---")
for token in tokens:
    print(f"{token} -> '{encoding.decode([token])}'")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Load a simple embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Define some words to compare
words = ["King", "Queen", "Apple", "Mango", "Man", "Woman"]

# 3. Convert words to vectors (Embeddings)
word_vectors = model.encode(words)

print(f"Vector Shape for '{words[0]}': {word_vectors[0].shape}")
print(f"First 5 numbers of '{words[0]}' vector: {word_vectors[0][:5]}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compare 'King' with others
def check_similarity(word1, word2_idx):
    sim = cosine_similarity([word_vectors[0]], [word_vectors[word2_idx]])[0][0]
    print(f"Similarity between '{words[0]}' and '{words[word2_idx]}': {sim:.4f}")

check_similarity("King", 1) # King vs Queen (High)
check_similarity("King", 2) # King vs Apple (Low)
check_similarity("King", 4) # King vs Man (High)
```
